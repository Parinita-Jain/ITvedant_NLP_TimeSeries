```
import asyncio
from rasa.core.agent import Agent
from rasa.utils.endpoints import read_endpoint_config

async def load_and_chat(model_path, endpoints_file):
    # Load endpoints
    endpoints = read_endpoint_config(endpoints_file, endpoint_type='str')

    # Initialize agent
    agent = Agent.load(model_path)

    # Ensure agent is ready
    if not agent.is_ready():
        raise Exception("Agent is not ready.")

    print("Agent is ready!")

# Example usage
model_path = model_path.model
endpoints_file = 'endpoints.yml'

# Run the loading and chatting process
loop = asyncio.get_event_loop()
loop.run_until_complete(load_and_chat(model_path, endpoints_file))
```
